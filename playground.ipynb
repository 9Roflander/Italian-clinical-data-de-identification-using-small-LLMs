```python
import json
import csv
#write every sample in a new line in a file, one column for the input and one for the annotations
samples_path = "data_annotated.jsonl"
output_path = "samples.csv"

samples = []

with open(samples_path, 'r') as f:
    for line in f:
        samples.append(json.loads(line))

csv_writer = csv.writer(open(output_path, 'w'))

#write the header
csv_writer.writerow(["input", "annotations","annotations_number"])

for sample in samples:
    csv_writer.writerow([sample["input"], sample["annotations"], len(sample["annotations"])])
```

```python
for sample in samples:
    print(sample["annotations"])
```

```python
annotated_data_path = "Annotated clinical notes - samples.csv"
import csv
annotated_data = []

with open(annotated_data_path, 'r') as f:
    csv_reader = csv.reader(f)
    next(csv_reader)  # Skip header row
    for row in csv_reader:
        annotated_data.append({
            "text": row[0],
            "annotations_redacted": row[1]
        })

```

```python
import json
annotation_type_count = {}
for row in annotated_data:
    print(row)
    for annotation in json.loads(row["annotations_redacted"]):
        #print(annotation)
        #print(json.loads(row["annotations_redacted"]))
        if not annotation["type"] in annotation_type_count.keys():
            
            annotation_type_count[annotation["type"]] = 1
        else:
            annotation_type_count[annotation["type"]] += 1

print(annotation_type_count)
```
